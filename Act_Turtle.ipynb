```python
!pip3 install ColabTurtle
from ColabTurtle.Turtle import *
import ColabTurtle

grados = 360
'''lados = int(input("¿Cuántos lados quieres tu figura: ?"))
longitud = int(input("Longitud de cada lado: "))'''

lados = 9
longitud = 30


lados_grados = grados  / lados

initializeTurtle(initial_speed=5)

pencolor("purple")

for i in range(lados):
  pencolor("red")
  forward(150)
  left(lados_grados)
  for i in range(lados):
    pencolor("yellow")
    forward(100)
    left(lados_grados)
    for i in range(lados):
      pencolor("green")
      forward(50)
      left(lados_grados)
```
